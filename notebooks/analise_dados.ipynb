```python
df = pd.read_csv('..\\data\\processed\\CPGF_consolidado_final.csv', sep=';')
df.head()
```

### Criação de KPIS

Valor total gasto em sigilo X Gastos com portadores identificados \
Percentual de transparência desses gastos \
Qual ministério possui o maior volume de dinheiro gasto sigilosamente 



```python
filtro_sigilosos = df['cpf_portador'] == 'sigiloso'

total_sigiloso = df[filtro_sigilosos]['valor_transacao'].sum()
total_identificado = df[~filtro_sigilosos]['valor_transacao'].sum()

print(f"Gastos Sigilosos ou Saques: R$ {total_sigiloso:.2f}")
print(f"Gastos com CPF identificado: R$ {total_identificado:.2f}")
```

```python
total_transacao = df['valor_transacao'].sum()

print(f"Total Gasto: R$ {total_transacao:,.2f}")

percent_transp = (total_identificado / total_transacao) * 100 

print(f"Percentual de transparencia: {percent_transp:.2f}%")
```

```python
orgSup_gastos_sigilos = df[filtro_sigilosos].groupby(by='nome_orgao_superior')['valor_transacao'].sum().sort_values(ascending=False)

orgSup_gastos_sigilos.head()
```

Top 5 orgaos superiores que mais gastaram no total com e sem sigilo?

```python
orgSup_gastos_total = df.groupby(by='nome_orgao_superior')['valor_transacao'].sum().sort_values(ascending=False)

orgSup_gastos_total.head(5)
```

TOP 10 Órgãos Subordinados de gastos totais?

```python
orgSub_gastos_total = df.groupby(by='nome_orgao')['valor_transacao'].sum().sort_values(ascending=False)

orgSub_gastos_total.head(10)
```

```python
favorecido_faturamento = df.groupby(by='nome_favorecido')['valor_transacao'].sum().sort_values(ascending=False)

favorecido_faturamento.head(10)
```
